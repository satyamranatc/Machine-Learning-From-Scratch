# LinearRegression With Mutiple Variables

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
Df = pd.read_csv("../data.csv")
Df.head()
```

```python
x = Df[["condition", "bathrooms","bedrooms","sqft_living"]]
x
```

```python
y= Df["price"]
y
```

```python
model = LinearRegression()
```

```python
model.fit(x, y)
```

```python
model.predict([[3,1.50,3.0,1340]])
```
