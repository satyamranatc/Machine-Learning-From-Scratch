```python
# Distance Theory
import numpy as np 
  
def distance(p1, p2): 
    return np.sqrt(np.sum(np.power(p2-p1, 2))) #distance between two points 
p1 = np.array([2 ,2])   #coordinate x = 1, y = 1 
p2 = np.array([5, 5])   #coordinate x = 4, y = 4 
distance(p1, p2) 
```

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
# KNN Algorithm
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

feature_names = iris.feature_names
target_names = iris.target_names

print(feature_names, target_names)
```

```python
import matplotlib.pyplot as plt


# Plotting
plt.figure(figsize=(10, 6))

# Use different colors for different classes
colors = ['red', 'green', 'blue']
for i, color in enumerate(colors):
    subset = X[y == i]
    plt.scatter(subset[:, 0], subset[:, 1], color=color, label=target_names[i])

# Add labels and title
plt.xlabel(feature_names[0])  # Sepal length
plt.ylabel(feature_names[1])  # Sepal width
plt.title("Iris Dataset: Sepal Length vs Sepal Width")
plt.legend()
plt.show()
```

```python
x_train, y_train, x_test, y_test = train_test_split(X,y,test_size=0.2, random_state=2)
```

```python
KNN_Model = KNeighborsClassifier(n_neighbors=3)
```

```python
KNN_Model.fit(X,y)
```

```python
# setosa' 'versicolor' 'virginica'

def Predeict(*d):
    d = [d]
    return KNN_Model.predict(d)

Res = Predeict(5.1, 3.8, 1.5, 0.3)
if Res == 0:
    print('setosa')
elif Res == 1:
    print('versicolor')

else:
    print('virginica')

```
